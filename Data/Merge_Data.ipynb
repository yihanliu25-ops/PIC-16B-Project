```python
import pandas as pd
from fredapi import Fred
fred = Fred(api_key = '621a31e8c3e8dcea20a5c0db2bafb43c')
```

```python
unemployment = fred.get_series('UNRATE')
treasury_spread = fred.get_series('T10Y2Y')
median_income = fred.get_series('MEHOINUSA646N')
consumer_sentiment = fred.get_series('UMCSENT')
```

```python
macro_data = pd.DataFrame({
    'unemployment': unemployment.groupby(unemployment.index.year).mean(),
    'treasury_spread': treasury_spread.groupby(treasury_spread.index.year).mean(),
    'consumer_sentiment': consumer_sentiment.groupby(consumer_sentiment.index.year).mean(),
    'median_income':median_income.groupby(median_income.index.year).mean()
})

macro_data.index.name = 'Year'
macro_data = macro_data.reset_index()
```

```python
df = pd.read_csv('term_life_experience_study.csv')
```

```python
df_merge = df.merge(macro_data, left_on='Issue_Year', right_on='Year', how='left')
```
