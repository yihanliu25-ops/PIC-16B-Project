```python
import pandas as pd
```

```python
df = pd.read_csv('insurance_data.csv')
```

```python
print(f'{df['Policy_ID'].duplicated().sum()}')
df.shape
```

```python
df = df.drop_duplicates(subset='Policy_ID', keep='first')
df.shape
```

```python
df = df[(df['Issue_Year'] >= 1990) & (df['Issue_Year'] <= 2025)]
df = df[(df['Birth_Year'] < 2025)]
df = df[(df['Face_Amount'] > 0)]
df = df[(df['Issued_Age'] > 0)]
```

```python
df.shape
```

```python
df = df.dropna(subset=['Smoker_Status', 'BMI'])
```

```python
df.shape
```

```python
df['Is_Smoker'] = (df['Smoker_Status'] == 'Smoker').astype(int)
df['Is_Female'] = (df['Gender'] == 'F').astype(int)
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = pd.get_dummies(df, columns=['Payment_Method'], drop_first=True)
```

```python
payment_cols = [col for col in df.columns if col.startswith('Payment_Method_')]
df[payment_cols] = df[payment_cols].astype(int)
```

```python
df = df.drop(columns = ['Post_Term_Premium','Policy_Agent_ID','ZIP_Code','Policy_Anniversary_Month','Year'])
```

```python
df['Lapsed'] = (df['Policy_Status'] == 'Lapsed').astype(int)
```

```python
df = df.drop(columns=[
    'Policy_Status',
    'Smoker_Status',
    'Gender',
])
```

```python
df['Premium_ratio'] = df['Annual_Premium'] / (df['Face_Amount'])
```

```python
df = df.drop(columns=[
    'Policy_ID',
    'Lapse_Year',
    'Death_Year',
    'Renewed',
])
```
